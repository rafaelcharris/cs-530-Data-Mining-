## 1. Combine all csv files in 'seasonprojections.zip' into a single file.
### 1.1 Change the name of Int in QBs to IntThrown

## 1.2 Merge all data sets

```python
df_list = [pd.read_csv(csv_file) for csv_file in os.listdir()]
merged_df = pd.concat(df_list)
```

## 2. Add column TotalTackles, which is equal to Tackles + Assists

```python
merged_df = merged_df.assign(
    TotalTackles=lambda x: merged_df["Tack"] + merged_df["Asst"])
```

## 3. Make a scatter plot of projected TDs and IntThrown for QBs

```python
df_plot = merged_df.loc[merged_df["Position"] == "QB"]
# 3.1 Create one variable for the TDs (There were two)
df_plot = merged_df.assign(
    TD=lambda x: merged_df["Pass TDs"] + merged_df["Rush TDs"])
# 3.2 PLot the scatter plot
plt.scatter(df_plot["TD"], df_plot["IntThrown"], alpha=0.7)
plt.xlabel("TDs")
plt.ylabel("IntThrown")
plt.show()
```
